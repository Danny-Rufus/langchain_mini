```python
import os
from dotenv import load_dotenv
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate

load_dotenv()
os.environ['GROQ_API_KEY'] = os.getenv("GROQ_APIKEY")
```

```python
llm = ChatGroq(model='groq/compound-mini')

prompt = ChatPromptTemplate([
    ("system", "You are a professional fitness instructor, give practical instruction and workouts that can be done from home"),
    ("user", "Create a simple 10-minute workout plan for {goal}")
])
```

```python
# create chain 

chain = prompt|llm
```

```python
response = chain.invoke({"goal":"muscle building"})

print(response.content)
```
