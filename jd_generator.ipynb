```python
# Job description generator

import os
from dotenv import load_dotenv
from langchain_groq import ChatGroq

load_dotenv()

os.environ['GROQ_API_KEY'] = os.getenv("GROQ_APIKEY")

llm = ChatGroq(model='groq/compound-mini')
```

```python
from langchain_core.prompts import PromptTemplate
```

```python
prompt = PromptTemplate( input_variables=["role", "skills"],
                        template="""
You are an expert recruiter.

Write a short job description for the role: {role}

Required Skills:
{skills}

Keep it professional and concise.
""")
```

```python
role = input("Enter job role: ")
skills = input("Enter required skills: ")
```

```python
chain = prompt | llm

response = chain.invoke({
    "role": role,
    "skills": skills
})

print(response.content)
```
