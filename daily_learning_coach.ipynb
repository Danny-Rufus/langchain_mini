```python
import os 
from dotenv import load_dotenv
from langchain_groq import ChatGroq
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
load_dotenv()

os.environ['GROQ_API_KEY'] = os.getenv("GROQ_APIKEY")

llm = ChatGroq(model='groq/compound-mini')
```

```python
prompt = PromptTemplate(
    input_variables= ['topics', 'time'],
    template="""
    You are a learning coach
    
    Create a {time} and a learning plan for {topic}.
    
    Return:
1. Topic
2. Resource suggestion
3. Mini exercise
    """
)

outputparser = StrOutputParser()
```

```python
chain = prompt |llm | outputparser
```

```python
response = chain.invoke({'time':"30 minutes", "topic": "RAG"})

# print(response)

from IPython.display import Markdown, display
display(Markdown(response))
```
